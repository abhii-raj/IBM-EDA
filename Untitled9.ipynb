### Residuals
- diffrences between actual and predicted values of the outcome variable.

Approaches to calculating residuals:
- Standarized: residual devided by standard errors
- Deleted: residual from fitting model on all data excluding current observation.
- stundeized: Deleted residuals divided by residual standard error (based on all data, or all data excluding current observation.)

## Variable selection: Background
variable selection involves choosing the set of features to include in the model.

Variable must often be transformed before they can be included in models. In addition to log and polynomial transformation this can involve :    
- Encoding : converting non-numeric features to numeric featuers.
- Scaling: converting the scale of numeric data so they are comparable.
The appropiate method of encoding depdnds on the type if feature.

### Feature Encoding
Encoding is often app;ied to catogorical features, that take non-numeric values.
Two primiary types:
- Nominal: Categorical variables take values in unordered categories (e.g red, blue, Green, True ,False)
- Ordinal : categorical variables take values in ordered catogaries(e.g. High, Medium, Low)

There are several common approaches to encoding variables:
- Binary encoding: converts variables to either 0 or 1 and is suitable for variables that take two possible values ( e.g.  true false )

- One - hot  encoding : converts variables that take multiple values into binary (0,1) variables, one for each category. This creates several new variables.

- ordinal encoding: involves converting ordered catogaries to numerical value, ususally by creating one variable that takes integer equal to the number of catogaries (e.g. 0,1,2,3.... )

### Feature scaling: Approaches
There are many approaches to scaling features.
some of the more common approaches include:
- Standard scaling: convert features to standard normal variables (by subtracting the mean and deviding by the standard error).

- Min-Max scaling: converts variables in the (0,1) interval by mapping minimum values to 0 and maximum values to 1.
this type of scaling is sensitive to outliers.  

    (for example if all values are under 1-10 an outlier is at 10,000 it will just   compress all the values at a side from 1-10 )


- Robust scaling: is similar to min-max scaling but instead maps the interquartile range.( the 75th percentile value minus the 25th percentile value ) to (0,1).
This means the variable itself takes values outside of the (0,1) interval.
